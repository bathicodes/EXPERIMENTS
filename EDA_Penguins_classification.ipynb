### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

### Import dataset

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
x = dataset.drop('species', axis=1)
y = dataset['species']
```
